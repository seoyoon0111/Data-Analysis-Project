```python
import numpy as np
import pandas as pd
import folium
import json
import googlemaps
```

```python
crime_in_police=pd.read_csv('crime_in_seoul.csv', thousands=',', encoding='euc-kr')
```

```python
gmaps_key = "AIzaSyBPB3Kg8DXsM1hR81EGdc-X6w3QbRPdeb4"    # 구글 맵 api 개인 키(제거 쓰세요. )
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
# 경찰서의 이름 정보를 추출
station_name = []
for name in crime_in_police['관서명']:
    station_name.append('서울' + str(name[:-1]) + '경찰서')  
station_name
```

```python
# googlemaps를 사용하여 위에서 추출한 경찰서의 이름을 가지고 경찰서의 위치를 찾아낸다
station_address = []
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name, language='ko')
    station_address.append(tmp[0].get("formatted_address"))
    
    tmp_loc = tmp[0].get("geometry")
    
    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])
    
    print(name + '-->' + tmp[0].get("formatted_address"))
```

```python
# 경찰서를 각 구별로 묶어준다 - 위에서 추출한 경찰서의 주소 정보를 활용 station_address
gu_name = []
for name in station_address:
    tmp = name.split()    
    tmp_gu = [gu for gu in tmp if gu[-1] == '구'][0]    
    gu_name.append(tmp_gu)
crime_in_police['구'] = gu_name
crime_in_police
```

```python
crime_in_police.to_csv('crime_in_police.csv', sep=',', encoding='utf-8')
```

```python
# 앞의 crime_in_police 데이터 프레임에서 위도/경도 데이터 추가
crime_in_police['lat'] = station_lat
crime_in_police['lng'] = station_lng
col = ['살인 검거', '강도 검거', '강간 검거', '절도 검거', '폭력 검거']
tmp = crime_in_police[col] / crime_in_police[col].max()
crime_in_police['검거'] = np.sum(tmp, axis=1)
```

```python
# 서울시 경찰서별 검거율을 원의 크기로 지도에 표식화
map_arrest = folium.Map(location=[37.5502, 126.982], zoom_start=11)
for n in crime_in_police.index:
    folium.CircleMarker([crime_in_police['lat'][n], crime_in_police['lng'][n]],
                        radius = crime_in_police['검거'][n]*10,
                        fill=True, color='#3186cc', fill_color='#3186cc',  # color를 hex 값으로 줌
                        fill_opacity=0.2).add_to(map_arrest)
map_arrest
```

```python
# 구별로 범죄 현황 묶어주기
crime_in=pd.pivot_table(crime_in_police, index='구', aggfunc=np.sum)
```

```python
# 각 범죄마다 검거율을 구한다
crime_in['강간검거율'] = crime_in['강간 검거']/crime_in['강간 발생']*100
crime_in['강도검거율'] = crime_in['강도 검거']/crime_in['강도 발생']*100
crime_in['살인검거율'] = crime_in['살인 검거']/crime_in['살인 발생']*100
crime_in['절도검거율'] = crime_in['절도 검거']/crime_in['절도 발생']*100
crime_in['폭력검거율'] = crime_in['폭력 검거']/crime_in['폭력 발생']*100
```

```python
# 검거율을 구했으니 검거 열은 삭제
del crime_in['강간 검거']
del crime_in['강도 검거']
del crime_in['살인 검거']
del crime_in['절도 검거']
del crime_in['폭력 검거']
```

```python
geo_path = 'korea_geo.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
# 검거율이 100%이 넘어가는 데이터들이 있다
# 이런 경우에 100%을 최대 값으로 조정
con_list = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
for column in con_list:
    crime_in.loc[crime_in[column] > 100, column] = 100
crime_in.head()
```

```python
# columns 이름을 다시 설정해준다
crime_in.rename(columns = {'강간 발생':'강간',
                             '강도 발생':'강도',
                             '살인 발생':'살인',
                             '절도 발생':'절도',
                             '폭력 발생':'폭력'}, inplace=True)
crime_in.head()
```

```python
# 간단한 머신러닝 예습해보기 - 각 범죄의 검거 비율을 예측해본다
from sklearn import preprocessing

crime = ['강간', '강도', '살인', '절도', '폭력']

x = crime_in[crime].values
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
crime_in_ratio = pd.DataFrame(x_scaled, columns=crime, index=crime_in.index)

crime2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_in_ratio[crime2] = crime_in[crime2]
crime_in_ratio.head(10)
```
